```python
x = data.Education
y = data.Income
print (x.values.shape, y.values.shape)
```

## 顺序模型
有一个输入和一个输出，顺序搭建的网络

```python
model = tf.keras.Sequential()
```

```python
# Dense(神经元数，特征数)
model.add(tf.keras.layers.Dense(1, input_shape=(1,)))
```

## 模型参数
Param 2  f(x) = ax + b 2指的是模型生成的对应于a, b两个参数
(None, 1) None指的是不限制样本数，1指的是输出维度

```python
model.summary()
```

```python
# 模型编译
model.compile(
    optimizer='adam', # 梯度下降
    loss='mse' # mean square error 方差
)
```

```python
# epochs 训练次数
history = model.fit(x, y, epochs=500)
```

```python
pred_y = model.predict(x)
```

```python
model.predict(x=pd.Series([8.0010]))
```

```python
model.evaluate(x, y)
```
